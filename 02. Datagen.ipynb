```python
import PIL
import os
from PIL import Image
import glob
import random
import yaml
import tqdm
```

#### Create char image list

```python
counts = 100
char_im_paths = []
home = os.environ['HOME']
char_dirs = os.listdir(home + "/Storage/Datasets/ETL/ETL2")
for char_dir in char_dirs:
    dir_im_paths = glob.glob(home + "/Storage/Datasets/ETL/ETL2/" + char_dir + "/*.png")
    dir_im_count = len(dir_im_paths)
    for i in range(counts):
        char_im_paths.append(dir_im_paths[i % dir_im_count])
```

```python
len(char_im_paths)
```

```python
random.shuffle(char_im_paths)
```

#### Place characters on pages

```python
with open("classes_inverse.yaml", 'r') as file:
    char_codes = yaml.safe_load(file)['names']
```

```python
print(char_im_paths[0].split("/")[-2])
```

```python
char_codes[char_im_paths[0].split("/")[-2]]
```

```python
train_data_dir = home + "/Storage/Datasets/jp_ocr_YOLO/YOLO_train/dataset/"
def make_page(i):
    page = Image.new('RGB', size=(874, 1240), color='white') # A7 paper
    char_im = Image.open(char_im_paths[0])
    pos_y = 0
    open(train_data_dir + "labels/train/" + str(i) + ".txt", "w").close()
    while pos_y + char_im.size[1] < page.size[1]:
        if random.randrange(0, 100) < 10:
            pos_y += char_im.size[1]
            continue
        
        pos_x = 17
        while pos_x + char_im.size[0] < page.size[0]:
            if random.randrange(0, 100) < 10:
                pos_x += char_im.size[0]
                continue
            
            char_im = Image.open(char_im_paths[0])
            char_id = char_codes[char_im_paths[0].split("/")[-2]]
            with open(train_data_dir + "labels/train/" + str(i) + ".txt", "a") as labels_file:
                labels_file.write(f"{char_id} {(pos_x + char_im.size[0] / 2) / page.size[0]} {(pos_y + char_im.size[1] / 2) / page.size[1]} {char_im.size[0] / page.size[0]} {char_im.size[1] / page.size[1]}\n")
            page.paste(char_im, (pos_x, pos_y))
            pos_x += char_im.size[0]
            del char_im_paths[0]
            if len(char_im_paths) == 0:
                break
        pos_y += char_im.size[1]
    page.save(train_data_dir + "images/train/" + str(i) + ".png")
    
```
